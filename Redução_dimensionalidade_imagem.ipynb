```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread('filhote-gato-01.jpg')
imagem
```

```python
# Converter a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Exibir a imagem em tons de cinza
print("Imagem em tons de cinza:")
cv2_imshow(imagem_cinza)
```

```python
# Binarizar a imagem (preto e branco)
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Exibir a imagem binarizada
print("Imagem binarizada (preto e branco):")
cv2_imshow(imagem_binaria)
```

```python
# Salvar as imagens resultantes
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_binaria.jpg', imagem_binaria)
```
